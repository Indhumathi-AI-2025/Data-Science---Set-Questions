A company wants to analyze customer satisfaction survey data. How can Data Science concepts help to clean and analyze this data?
SampleDataset:
    'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]


```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]
    
})
data
```

```python
#STATISTICS
data.describe()
```

```python
data['Satisfaction'] = data['Satisfaction'].fillna(data['Satisfaction'].mean()).round(1)
data
```

```python
#DATA VISUALIZATION
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(data['Satisfaction'])
plt.title("The Customer's Satisfaction Survey Report")
plt.show()
```

Question 2 :
 A retail store has recorded monthly sales data for its products. How can Data Science concepts be applied to understand sales performance?
SampleDataset:
'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
 'Sales': [1500, 1700, 1600, 1800, 2200, 2100]


```python
#The retail store has an average sale of rupees 1817 per month; some months(May and June) have more sales than the average sale rate. 
#And, jan, feb & Mar has the minimum sale. Overall, May has the highest sales.
```

```python
import pandas as pd
import numpy as np

retail_data = pd.DataFrame({'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
 'Sales': [1500, 1700, 1600, 1800, 2200, 2100]
    
})
retail_data
```

```python
#Sales Performance

retail_data_mean = retail_data['Sales'].fillna(retail_data['Sales']).mean().round(0)
retail_data_mean
```

```python
avg_sales = retail_data['Sales'].mean().round(0)
max_sales= retail_data['Sales'].max().round(0)
min_sales = retail_data['Sales'].min().round(0)
print("The Retail Store's:\nAverage Sales is:",avg_sales)
print("Maximum Sales is:",max_sales )
print("Minimum Sales is:",min_sales )
```

```python
max_sales_index = retail_data['Sales'].idxmax()
highest_sales_month = retail_data.loc[max_sales_index, 'Month']
print(f"The month with the highest sales is: {highest_sales_month}")
```

```python
#DATA VISUALIZATION
# Create the line plot using seaborn
sns.lineplot(x='Month', y='Sales', data=retail_data)

# Add a title
plt.title("The Retail Store's Monthly Sales Trend")

# Display the plot
plt.show()
```

A hospital conducted a study to check the effectiveness of a new medical test. How can Data Science concepts be used to validate its results?
SampleDataset:
group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]
Expected Output:
T-Statistic: -5.92, P-Value: 0.0004
Reject Null Hypothesis: The test is effective.


```python
import pandas as pd
import numpy as np
from scipy import stats

group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]

t_stat, p = stats.ttest_ind(group_a, group_b)
t_stat, p
print("T-Statistic: ",t_stat )
print("P-Value: ",p )
if(p<0.05):
    print("Reject Null Hypothesis: The test is effective.")
else:
    print("")
```

 A company wants to evaluate employee performance using various metrics. How can Data Science concepts be applied?
Sample Dataset:
    'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]

Perform:
1)	Data Cleaning
2)	Descriptive Statistics
3)	Correlation Analysis (Using heatmap)
4)	Distribution Analysis (Plot Histogram)
5)	Performance Comparison (Using Boxplot)
6)	Experience vs Performance (Perform Bivariate Analysis using a scatter plot)
7)	Department-wise Analysis


```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]
})
```

```python
df
```

```python
#1.Data Cleaning

df['Department'].fillna('Unknown', inplace = True)
df['Experience (Years)'].fillna(df['Experience (Years)'].mean(), inplace = True)
df['Projects Completed'].fillna('0', inplace = True)
df['Performance Rating'].fillna(df['Performance Rating'].mean(), inplace = True)
df
```

```python
#2)	Descriptive Statistics

df.describe()
```

```python
#3)	Correlation Analysis (Using heatmap)

import matplotlib.pyplot as plt
import seaborn as sns

df_corr = df.corr(numeric_only = True)
plt.figure(figsize=(10,8))
sns.heatmap(df_corr, annot = True, cmap='coolwarm')
plt.show()
```

```python
#4)	Distribution Analysis (Plot Histogram)


import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Performance Rating'])
plt.show()
```

```python
#5)	Performance Comparison (Using Boxplot)

plt.title("Department wise - Projetcs Completed")
sns.boxplot(x='Department', y='Projects Completed', data=df, color = 'green')
plt.show()
```

```python
#6)	Experience vs Performance (Perform Bivariate Analysis using a scatter plot)

from scipy import stats
from scipy.stats import f_oneway

t_stat, p_value = stats.f_oneway(df['Experience (Years)'], df['Performance Rating'])
t_stat, p_value
```

```python
sns.scatterplot(x=df['Experience (Years)'] , y =df['Performance Rating'], color = 'Orange')
plt.show()
```

```python
#7)	Department-wise Analysis

department_analysis = df.groupby('Department').agg(
    AverageProjects=('Projects Completed', 'mean'),
    AveragePerformance=('Performance Rating', 'mean'),
    EmployeeCount=('Department', 'count')
).reset_index()

print("\nDepartment-wise Analysis Results:")
print(department_analysis)
```
