A company wants to analyze customer satisfaction survey data. How can Data Science concepts help to clean and analyze this data?
SampleDataset:
    'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]


```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]
    
})
data
```

```python
data['Satisfaction'] = data['Satisfaction'].mean().round(1)
data['Satisfaction']
data
```

A hospital conducted a study to check the effectiveness of a new medical test. How can Data Science concepts be used to validate its results?
SampleDataset:
group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]
Expected Output:
T-Statistic: -5.92, P-Value: 0.0004
Reject Null Hypothesis: The test is effective.


```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway

group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]

f_stat, p = f_oneway(group_a,group_b )
f_stat, p
print("T-Statistic: ",f_stat )
print("P-Value: ",p )
if(p<0.05):
    print("Reject Null Hypothesis: The test is effective.")
else:
    print("")
```

 A company wants to evaluate employee performance using various metrics. How can Data Science concepts be applied?
Sample Dataset:
    'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]

Perform:
1)	Data Cleaning
2)	Descriptive Statistics
3)	Correlation Analysis (Using heatmap)
4)	Distribution Analysis (Plot Histogram)
5)	Performance Comparison (Using Boxplot)
6)	Experience vs Performance (Perform Bivariate Analysis using a scatter plot)
7)	Department-wise Analysis


```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]
})
```

```python
df
```

```python
#1.Data Cleaning

df['Department'].fillna('Unknown', inplace = True)
df['Experience (Years)'].fillna(df['Experience (Years)'].mean(), inplace = True)
df['Projects Completed'].fillna('0', inplace = True)
df['Performance Rating'].fillna(df['Performance Rating'].mean(), inplace = True)
df
```

```python
df.drop(columns='Dept_Max_Exp')
```

```python
#2)	Descriptive Statistics
#There are 3 departments in the company. The employee who has more experience has done more projects 
#and achieved a good performance rating.

df['Department'].value_counts()
```

```python
#3)	Correlation Analysis (Using heatmap)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(df, annot = True, cmap='blue')
plt.show()
```

```python
#4)	Distribution Analysis (Plot Histogram)


import matplotlib.pyplot as plt
import seaborn as sns

plt.hist(x=df['Department'], y=df['Performance Rating'], data = df)
plt.show()
```

```python
#5)	Performance Comparison (Using Boxplot)

plt.boxplot(df['Performance Rating'])
plt.show()
```

```python
#6)	Experience vs Performance (Perform Bivariate Analysis using a scatter plot)

from scipy import stats
from scipy.stats import f_oneway

t_stat, p_value = stats.f_oneway(df['Experience (Years)'], df['Performance Rating'])
t_stat, p_value
```

```python
sns.scatterplot(x=df['Experience (Years)'] , y =df['Performance Rating'])
plt.show()
```
